## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from google.colab import files
uploaded = files.upload()

```

```python
#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv('charity_data.csv')
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df = application_df.drop(['EIN','NAME'], axis = 1)
#  YOUR CODE GOES HERE
```

```python
df.columns.unique()
```

```python
df.info()
```

```python
df.nunique()
```

```python
df['ASK_AMT'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
counts = df['APPLICATION_TYPE'].value_counts()
counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
application_types_to_replace= list(counts[counts<500].index)  

application_types_to_replace
```

```python
# Replace in dataframe
for app in application_types_to_replace:
   df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
count_classi = df['CLASSIFICATION'].value_counts()
count_classi
```

```python
# counts_classification = count_classi[count_classi>1]
# counts_classification
```

```python
classification_types_to_places1 = list(count_classi[count_classi<100].index)
classification_types_to_places1 
```

```python
for cls in classification_types_to_places1:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls, 'Others')
  
```

```python
df['CLASSIFICATION'].value_counts()
```

```python
df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
df = pd.get_dummies(df,dtype=float)
df.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = df['IS_SUCCESSFUL'].values
y
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
```

```python
X = df.drop(['IS_SUCCESSFUL'],axis=1)
X.head()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
num_input_features= len(X_train_scaled[0])
hidden_node_layer1 = 7
hidden_node_layer2 =14
hidden_node_layer3 = 21

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units= hidden_node_layer1, input_dim=num_input_features , activation='relu'))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_node_layer2, activation='relu'))
# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy' , optimizer='adam', metrics=['accuracy'])

#  YOUR CODE GOES HERE
```

```python
# Train the model
fit_model= nn.fit(X_train_scaled, y_train, validation_split=0.15, epochs=100)
#  YOUR CODE GOES HERE
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
